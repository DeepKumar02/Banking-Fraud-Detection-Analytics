```python
import pandas as pd
import numpy as np
```

```python
# importing datasets
users = pd.read_csv("../Dataset/Raw_Data/users.csv")
cards = pd.read_csv("../Dataset/Raw_Data/cards.csv")
transactions = pd.read_csv("../Dataset/Raw_Data/credit_card_transactions.csv")
```

```python
# checking currency column (users, cards, transaction)
users[['Per Capita Income - Zipcode',
       'Yearly Income - Person',
       'Total Debt']].head()
```

```python
cards['Credit Limit'].head()
```

```python
transactions['Amount'].head()
```

```python
# convert currency column from objects to float64 (We'll remove $ and commas)
# users
currency_columns = [
    'Per Capita Income - Zipcode',
    'Yearly Income - Person',
    'Total Debt'
]

for col in currency_columns:
    if users[col].dtype == 'object':
        users[col] = (
            users[col]
            .str.replace('$', '', regex=False)
            .str.replace(',', '', regex=False)
            .astype(float)
        )
```

```python
# cards
if cards['Credit Limit'].dtype == 'object':
    cards['Credit Limit'] = (
        cards['Credit Limit']
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .astype(float)
    )
```

```python
if transactions['Amount'].dtype == 'object':
    transactions['Amount'] = (
        transactions['Amount']
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .astype(float)
    )
```
